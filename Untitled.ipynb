# Try surface model.

First install numpy and matplotlib, using `pip3 install numpy` and `pip3 install matplotlib`.

```python
# Surface model, assuming constant rate.
# We solve the following equations:
# k_s \phi = 1 / S_g / M * dX / dt
# - d\phi/dt = k_p k_s \phi^n

import numpy as np
from matplotlib import pyplot as plt

n = 1.7

# All parameters can be adjusted
kp, ks, Sg, M = 0.1, 0.1, 1, 1

phi0 = 1 # active surface fraction at t=0
dt = 1e-2 # numerically solve with time interval = 1e-4
totalt = 2000 # Reaction in the first 2000 seconds
nsteps = int(totalt/dt) # total number of steps in numerical calculation
t = np.linspace(0, totalt, nsteps+1)

phi = [phi0]
for istep in range(nsteps):
    phi_next = phi[-1] - kp * ks * phi[-1]**n * dt
    phi.append(phi_next)

plt.plot(t, phi)
plt.xlabel('Time (t)')
plt.ylabel('Active surface fraction')
plt.show()
```

```python
X = [0] # initial production quantity
for istep in range(nsteps):
    X.append(X[-1] + ks * phi[istep] * Sg * M * dt)

plt.plot(t, X)
plt.xlabel('Time (t)')
plt.ylabel('Total production')
plt.show()
#print(X)
```
